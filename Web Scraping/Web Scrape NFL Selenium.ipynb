```python
from selenium import webdriver
import pandas as pd
```

```python
browser = webdriver.Chrome("chromedriver.exe")
```

```python
url = "https://projects.fivethirtyeight.com/2017-nfl-predictions/"
browser.get(url)
```

```python
weeks = browser.find_elements_by_css_selector("option")
filters = [w.text for w in weeks]
```

```python
teams = []
mps = []
wds = []
frbs = []
wsbs = []
week = []
```

```python
# Tag name change after i=4
for i in range(5):
    weeks[i].click()
    
    team = browser.find_elements_by_css_selector("span.full")
    team = [t.text for t in team]
    teams = teams + team
    
    details = browser.find_elements_by_css_selector("td.pct")
    details = [d.text for d in details]
    
    mp = []
    wd = []
    frb = []
    wsb = []
    
    for j in range(0,len(details),4):
        mp.append(details[j])
        wd.append(details[j+1])
        frb.append(details[j+2])
        wsb.append(details[j+3])
   
    mps = mps + mp
    wds = wds + wd
    frbs = frbs + frb
    wsbs = wsbs + wsb
    
    wk = []
    wk.append(filters[i])
    week = week + wk*len(team)    
    
```

```python
for i in range(5,len(weeks)):
    weeks[i].click()
    
    team = browser.find_elements_by_css_selector("span.full")
    team = [t.text for t in team]
    teams = teams + team
    
    mp = browser.find_elements_by_css_selector("td.pct.div")
    mp = [m.text for m in mp]
    mps = mps + mp
    
    wd = browser.find_elements_by_css_selector("td.pct.division-chances")
    wd = [w.text for w in wd]
    wds = wds + wd
    
    frb = browser.find_elements_by_css_selector("td.pct.desktop")
    frb = [f.text for f in frb]
    frbs = frbs + frb
    
    wsb = browser.find_elements_by_css_selector("td.pct.superbowl")
    wsb = [w.text for w in wsb]
    wsbs = wsbs + wsb
    
    wk = []
    wk.append(filters[i])
    week = week + wk*len(team)
    
```

```python
for i in range(len(teams)):
    if wds[i] == '✓':
        wds[i] = '100%'
    elif wds[i] == '—':
        wds[i] = '0%'
    else:
        continue
```

```python
for i in range(len(teams)):
    if mps[i] == '✓':
        mps[i] = '100%'
    elif mps[i] == '—':
        mps[i] = '0%'
    else:
        continue
```

```python
for i in range(len(teams)):
    if frbs[i] == '✓':
        frbs[i] = '100%'
    elif frbs[i] == '—':
        frbs[i] = '0%' 
    else:
        continue
```

```python
for i in range(len(teams)):
    if wsbs[i] == '✓':
        wsbs[i] = '100%'
    elif wsbs[i] == '—':
        wsbs[i] = '0%'
    else:
        continue
```

```python
df = pd.DataFrame(data={"Week": week, 
                        "Teams": teams, 
                        "Make PlayOffs": mps, 
                        "Win Division": wds, 
                        "1st Round Bye": frbs,
                        "Win Super Bowl": wsbs})
df.groupby(['Week'])
```
