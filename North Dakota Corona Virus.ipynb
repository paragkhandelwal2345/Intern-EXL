```python
from selenium import webdriver
import pandas as pd
```

```python
browser = webdriver.Chrome("chromedriver.exe")
```

```python
browser.get("https://www.health.nd.gov/diseases-conditions/coronavirus/north-dakota-coronavirus-cases")
```

```python
# sources of exposure using selenium
graph_points = browser.find_elements_by_css_selector("tspan.highcharts-text-outline")
graph_points = [g.text for g in graph_points]
exposure = []
```

```python
for i in range(372,378):
    exposure.append(graph_points[i])
```

```python
# sources of exposure using csv file from api method
data = pd.read_csv("49s5b.csv")
data.loc['Total',:] = data.sum(axis=0)
```
